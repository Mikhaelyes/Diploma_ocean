```python
# gnome can use either ISO style strings or datetime objects
start_time="2025-08-29T11:28:00"
model = gs.Model(start_time="2025-08-29T11:28:00",
                 duration=gs.days(4),
                 time_step=gs.minutes(30)
                 )

# The base GnomeMap is all water, no land
# you can optionally add boundaries
model.map = gs.GnomeMap(map_bounds=((35, 43), (35, 47),
                                    (39, 47), (39, 43))
                        )

images_dir = './output'
netcdf_file = os.path.join(images_dir, 'test_output_v2.nc')
gs.remove_netcdf(netcdf_file)



# file_list = ['Data/Glorys/mercatorglorys12v1_gl12_mean_20250828_R20250903.nc',
#             'Data/Glorys/mercatorglorys12v1_gl12_mean_20250829_R20250903.nc',
#              'Data/Glorys/mercatorglorys12v1_gl12_mean_20250830_R20250903.nc',
#              'Data/Glorys/mercatorglorys12v1_gl12_mean_20250831_R20250903.nc',
#              'Data/Glorys/mercatorglorys12v1_gl12_mean_20250901_R20250903.nc',
#              'Data/Glorys/mercatorglorys12v1_gl12_mean_20250902_R20250903.nc'
#              ]

file_list = ['Data/Glorys/mercatorglorys12v1_gl12_mean_20250828_R20250903.nc']

current = gs.GridCurrent.from_netCDF(file_list)
current_mover = gs.CurrentMover(current=current)
model.movers += current_mover



file_wind_list = ['Data/Wind/b3300ed84673498a28c611d70bf51d40.nc']

wind = gs.GridWind.from_netCDF(file_wind_list)
wind_mover = gs.WindMover(wind)
model.movers += wind_mover

waves = gnome.environment.Waves(wind)
water = gnome.environment.Water(temperature=300.0, salinity=18.0)
model.weatherers += gnome.weatherers.NaturalDispersion(waves=waves,water=water)



# create spill
spill = gs.point_line_spill(release_time="2025-08-29T11:28:00",
                                    start_position=(37.637, 44.621),
                                    end_position=(37.404, 44.690),
                                    amount=36,
                                    units='bbls',
                                    end_release_time= "2025-08-29T18:19:00",
                                    substance=gs.GnomeOil(filename="alaska-north-slope_AD00020.json"),
                                    num_elements=500)
# add it to the model
model.spills += spill
print(model.spills)

# release = gs.PointLineRelease(num_elements=50,
#                               release_time="2025-08-29T11:28:00",
#                               start_position=(37.449, 45.695, 0),
#                               )

# spill = gs.Spill(release=release,
#                  substance=gs.GnomeOil(filename="alaska-north-slope_AD00020.json"),
#                  amount=36,
#                  units='bbls')

# model.spills += spill


# create an outputter: this renders png files and an animated gif
renderer = gs.Renderer(output_dir='./output/',
                       output_timestep=gs.hours(3),
                       # bounding box for the output images
                       viewport=((35, 43), (35, 47),
                                 (39, 47), (39, 43)),
                       formats=['gif']
                       )

model.outputters += renderer


nc_outputter = gs.NetCDFOutput(netcdf_file,
                               which_data='most',
                               output_timestep=gs.hours(6)
                               )
model.outputters += nc_outputter



print("running the model: see output in the output dir")
# run the model
model.full_run()
```
